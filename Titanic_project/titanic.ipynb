```python
#1.load the data and display it
import pandas as pd

# Load the Titanic dataset
df = pd.read_csv("train.csv")

# Display first 5 rows
df.head()
```

```python
#2.check for missing values
df.isnull().sum()
```

```python
#3.Check Data Types of Each Column
print(df.dtypes)
```

```python
# 4.Handle Missing Values
df['Age'] = df['Age'].fillna(df['Age'].median())
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
df.head()
```

```python
# 5.Print Column Names
print(df.columns)
```

```python
# 6.Convert categorical features to numerical for ml model
df = df.copy()  # Create a copy to avoid chained assignment issues
df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})
df.head()
```

```python
# 7.Data Visualization
import seaborn as sns
import matplotlib.pyplot as plt

# Countplot for survival
sns.countplot(x='Survived', data=df)
plt.show()

# Survival by Passenger Class
sns.barplot(x="Pclass", y="Survived", data=df)
plt.show()
```

```python
#8.Drop Unnecessary Columns
df = df.drop(['Name', 'Ticket', 'Cabin'], axis=1)
print(df.columns)
plt.show()

```
